```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import pip

tmbscorefile = "/Users/kogantit/Documents/TMB/PNOC003_analysis/all_PONOC003.tmb.txt"

#########################################################################
#########  Preparing input file #########################################

# Reading input file into pandas
tmb_scores_df = pd.read_csv(tmbscorefile, sep="\t")



median_width = 0.6
# Defining figure size
plt.figure(figsize=(20, 10))
# This plots all the disease types
ax = sns.stripplot(
    x="project",
    y="TMB",
    data=tmb_scores_df,
    jitter=1,
    linewidth=0.5,
)

# Taking log scale of y-axis
plt.yscale("log")
plt.xticks(fontsize=15, rotation=90)
ax.set_title("TMB scores", fontsize=30)  # title of plot
ax.set_xlabel("Disease type", fontsize=20)  # xlabel
ax.set_ylabel(
    "Mutations per Mb", fontsize=20
)  # ylabelplt.savefig('results/pbta-snv-mutect2.TMB.png')
ax.axhline(1, ls="dotted", color="gray")
ax.axhline(2, ls="dotted", color="gray")
ax.axhline(10, ls="dotted", color="gray")
ax.axhline(100, ls="dotted", color="gray")
plt.tight_layout()
plt.savefig("../output/PNOC003_tmb_WXS_and_WGS.png")

```
