```python
final = []
names = ['A','B','C','D']
for i,m in enumerate([1.5,1.2,1.6,1.8]):
    mser = pd.DataFrame()
    x = np.random.normal(m,.6,100)
    x = 10**x
    mser['TMD'] = x
    mser['Name'] = names[i]
    final.append(mser)
final = pd.concat(final)
final.index = np.arange(len(final))
```

```python
final = final.sample(frac=1)
final_gr = final.groupby('Name')
names = final['Name'].unique()
colors = ['b','g','r','c']
```

```python
plt.figure(figsize = (14,6))
for i,name in enumerate(names):
    plt.subplot(1,len(names),i+1)
    x = final_gr.get_group(name)
    x = x.sort_values('TMD')
    x.index = np.arange(len(x))/len(x)
    x['TMD'].plot(marker = 'o',ms=5,mfc='w',lw=0,c=colors[i])
    plt.yscale('log')
    plt.ylim(.1,3000)
    plt.yticks([.1,0.2,1,2,10,20,100,200,1000])
    plt.xticks([])
    ax=plt.gca()
    ax.set_yticklabels(['0.1','0.2','1','2','10','20','100','200','1000'])
    plt.hlines(x['TMD'].median(),-.05,1.05)
    plt.xlim(-.05,1.05)
    plt.title(name)
```
